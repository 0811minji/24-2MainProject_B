# Send_Keys

```python
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.by import By

def crawl_reviews(url):        
    driver = webdriver.Chrome()

    # 입력받은 URL 열기
    driver.get(url)

    # 페이지 로딩 대기
    time.sleep(2)

    # 리뷰 탭 클릭
    review_tab = driver.find_element(By.CSS_SELECTOR, '#QA0Szd > div > div > div.w6VYqd > div.bJzME.tTVLSc > div > div.e07Vkf.kA9KIf > div > div > div:nth-child(3) > div > div > button:nth-child(2) > div.LRkQ2 > div.Gpq6kf.fontTitleSmall')
    review_tab.click()

    driver.implicitly_wait(5)

    data = {
        'ID': [],
        'Review': [],
        'Rating': [],
        'Time': []
    }

    body = driver.find_element(By.CSS_SELECTOR, "#QA0Szd > div > div > div.w6VYqd > div.bJzME.tTVLSc > div > div.e07Vkf.kA9KIf > div > div > div.m6QErb.DxyBCb.kA9KIf.dS8AEf.XiKgde")
    for i in range(150):
        body.send_keys(Keys.PAGE_DOWN)
        time.sleep(1)

    # 중복 추적용 set
    seen_reviews = set()

    # 리뷰 상자 수집
    review_boxes = driver.find_elements(By.CSS_SELECTOR, 'div.jftiEf')

    # 각 리뷰 상자에서 데이터 추출
    for box in review_boxes:
        try:
            user_id = box.find_element(By.CSS_SELECTOR, 'button > div.d4r55').text
        except:
            user_id = None

        try:
            review_text = box.find_element(By.CSS_SELECTOR, 'span.wiI7pd').text
        except:
            review_text = None

        try:
            rating = box.find_element(By.CSS_SELECTOR, 'span.kvMYJc').get_attribute('aria-label')
        except:
            rating = None

        try:
            review_time = box.find_element(By.CSS_SELECTOR, 'span.rsqaWe').text
        except:
            review_time = None

        # 리뷰 텍스트가 공백이 아닌 경우만 수집
        if review_text and review_text.strip():  # 공백 체크 추가
            # 중복 확인 (리뷰 텍스트와 사용자 ID 조합)
            if (user_id, review_text) not in seen_reviews:
                seen_reviews.add((user_id, review_text))  # 중복 방지용 추가
                data['ID'].append(user_id)
                data['Review'].append(review_text)
                data['Rating'].append(rating)
                data['Time'].append(review_time)

    # 데이터프레임 생성
    df = pd.DataFrame(data)

    # 드라이버 종료
    driver.quit()
    
    return df
```

```python
data = pd.read_excel("추가 크롤링.xlsx")
for i, row in data.iterrows():
        name = row['name']  # 식당 이름
        url = row['url']
        try:
            df = crawl_reviews(url)
            df.to_excel(f"{name}.xlsx", index=False)
        except Exception as e:
            print(f"오류 발생 - 식당 이름: {name}, 오류 메시지: {e}")
```
